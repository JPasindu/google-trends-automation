```python
import pandas as pd
```

```python
COUNTRIES_FILE='countries.txt'
countries = []
with open(COUNTRIES_FILE, 'r', encoding='utf-8') as f:
    for line in f:
        line = line.strip()
        countries.append(line)
```

```python
for i in countries:
    file=f'google_trends_master_{i}.csv'
    df=pd.read_csv(file)
    if df.shape[1]<40:
        print(file)
```
